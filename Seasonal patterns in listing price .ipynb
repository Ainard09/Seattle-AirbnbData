# Data Understanding

## Access and explore the data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#read csv files from machine
listings_df = pd.read_csv('Seattle.listings.csv')
calendar_df = pd.read_csv('Seattle.calendar.csv')
reviews_df = pd.read_csv('Seattle.reviews.csv')

listings_df.head()
```

```python
print(f"rows {listings_df.shape[0]} and columns {listings_df.shape[1]}")
listings_df.describe()
```

```python
# listing_id
calendar_df['listing_id']= calendar_df['listing_id'].astype(str)

# create column for month_year, month and year
calendar_df['month'] = pd.DatetimeIndex(calendar_df['date']).month
calendar_df['year'] = pd.DatetimeIndex(calendar_df['date']).year
calendar_df['month_year'] = pd.to_datetime(calendar_df['date']).dt.to_period('M')


# remove the $ sign of the price on calendar table
calendar_df['price'] = calendar_df['price'].apply(lambda x: float((x[1:].replace(',',''))) if type(x) != float else x)

calendar_df.head()
```

```python
# convert t to 1 and f to 0 on available column
calendar_df["available"]=calendar_df["available"].apply(lambda x:1 if x== "t" else 0)
calendar_df.head()
```

```python
# store unbooked listing in calendar_booked
calendar_unbooked= calendar_df[calendar_df['available']== 1]
calendar_unbooked

# compute unbookings in each month
avg_month= calendar_unbooked['month_year'].value_counts().sort_values().reset_index()
avg_month.rename(columns={"index":"month_year","month_year":"available_listing"},inplace=True)
avg_month
```

```python
# store booked listing in calendar_booked
calendar_booked= calendar_df[calendar_df['available']== 0]
calendar_booked

# compute bookings in each month
avg= calendar_booked['month_year'].value_counts().sort_values().reset_index()
avg
```

```python
# Plot relationship between average price and number of available listings
fig,axis = plt.subplots()
axis.plot(calendar_df[calendar_df['year'] == 2016].groupby(['month'])['price'].mean(),color='green', marker='o', linewidth=3, label='Price')
axis.set_xlabel('month')
axis.set_ylabel('Avg Price, $')

axis2 = axis.twinx()
axis2.plot(calendar_df[calendar_df['year'] == 2016].groupby(['month'])['available'].sum(), color='violet', marker ='o', linewidth=3, label='Listing')
axis2.set_ylabel('No of Available Listings ')

axis.legend(loc=2)
axis2.legend(loc=4)

plt.title('Relationship Between Price and No of Available Listings')
plt.show()
```
